# Merging the datasets related to the Ufal Informatics' thesis

## 1. Importing the required libraries

```python
# Importing the required libraries.
import pandas as pd, numpy as np, csv
```

## 2. Getting the data

```python
# Getting the data.
df1 = pd.read_csv("../data/ufal_main_campus_thesis.csv", header=0, index_col=False, dtype=object)
df2 = pd.read_csv("../data/ufal_arapiraca_campus_thesis.csv", header=0, index_col=False, dtype=object)
```

```python
# Showing the information about the datasets.
print(f"{df1.info()}\n\n{df2.info()}")
```

```python
# Showing the first five records.
df1.head()
```

```python
# Showing the first five records.
df2.head()
```

## 3. Merging the datasets

```python
# Handling the nullable values.
df1.replace({np.nan: None}, inplace=True)
df2.replace({np.nan: None}, inplace=True)
```

```python
# Merging the datasets.
df = pd.concat([df1, df2], ignore_index=True)
```

```python
# Showing the result after the merge process.
df.head()
```

```python
# Showing the information about the data.
df.info()
```

## 4. Saving the final dataset

```python
# Saving the final dataset.
df.to_csv("../data/ufal_informatics_thesis.csv", index=False, quoting=csv.QUOTE_ALL)
```
